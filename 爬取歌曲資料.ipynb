```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
```

```python
headers = {'content-type': 'text/html; charset=UTF-8', 
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'}
```

```python
class Song:
    def __init__(self, name, ID):
        self.ID = ID
        self.name = name
        self.link = f'http://sekai.best/music/{self.ID}'
        self.difficulties = [-1,-1,-1,-1,-1]
        
    def __str__(self):
        return f"ID: {self.ID}, Name: {self.name}"
    
    def get(self):
        return f"Difficulty: {self.difficulties}\nlink: {self.link}"
```

```python
url = 'https://sekai-world.github.io/sekai-master-db-diff/musics.json'
res = requests.get(url, headers=headers)
pre_url = 'https://sekai.best/music/'
songs = {}
Json = eval(res.text)
for data in Json:
    songs[data['title']] = Song(data['title'], data['id'])
    print(songs[data['title']])
```

```python
List_url = 'https://projectsekai.fandom.com/wiki/Project_SEKAI_COLORFUL_STAGE!/Song_List'
List_res = requests.get(List_url, headers=headers)
List_soup = BeautifulSoup(List_res.text, 'html.parser')
song_sel = List_soup.select('table.wikitable tr td a')
for song in song_sel:
    if 'fandom' in song['href'] or 'twitter' in song['href']:
        continue
    song_name = song['title']
    song_url = 'https://projectsekai.fandom.com' + song['href']
    song_res = requests.get(song_url, headers=headers)
    song_soup = BeautifulSoup(song_res.text, 'html.parser')
    sel = song_soup.select('td span.hidden')
    if song_name in songs:
        curr_song = songs[song_name]
    elif song_soup.select('div.pi-data-value.pi-font')[0].text in songs:
        curr_song = songs[song_soup.select('div.pi-data-value.pi-font')[0].text]
    else:
        print(song_name, "not in songs.")
        continue
    for i in range(5):
        curr_song.difficulties[i] = int(sel[i].text.split()[1])
    print(curr_song, ", end.")
    time.sleep(10)
```

```python
for song_name, song_obj in songs.items():
    print(song_obj, song_obj.difficulties)
```
